**Step 1:** Randomly select 20 image out of the "tiles2" folder and store them in a folder called "tiles_keep".

```python
# Install packages
import os
import random
import shutil

# Set up
source = "/content/drive/My Drive/tiles2"
dest = "/content/drive/My Drive/tiles_keep"
no_of_files = 20

# Create destination folder if it doesn't exist
os.makedirs(dest, exist_ok=True)

# Get all TIFF files from the source folder
files = [
    f for f in os.listdir(source)
    if f.lower().endswith((".tif", ".tiff"))
]

# Safety check in case there are no tiff files on the folder
if no_of_files > len(files):
    raise ValueError(f"Only {len(files)} TIFF files found, cannot select {no_of_files}")

# Random selection of the tiff files in the folder
selected_files = random.sample(files, no_of_files)

print("Moved files:\n")

for i, file in enumerate(selected_files, start=1):
    source_file = os.path.join(source, file)
    dest_file = os.path.join(dest, file)

    shutil.move(source_file, dest_file)
    print(f"{i}) {file}")

print("\nDone.")
```
